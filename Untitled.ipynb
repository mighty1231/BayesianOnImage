```python
import pymc3 as pm
from theano.compile.ops import as_op
import theano.tensor as tt
import numpy as np
import matplotlib.pyplot as plt
import cv2
import random
import scipy 

from skimage.measure import compare_ssim as ssim
```

```python
# 여기부터는 color-version. 즉, gray-scale이 아닌 rgb color로 해보았습니다.

sample2 = cv2.imread("sample/starry_night.png", cv2.IMREAD_COLOR)
h, w, c = sample2.shape
sample2 = cv2.cvtColor(sample2, cv2.COLOR_BGR2RGB)
plt.imshow(sample2)
print(sample2.shape)
```

```python
@as_op(itypes=[tt.lvector, tt.lvector, tt.lvector, tt.lvector, tt.lmatrix], otypes=[tt.lscalar])
def compute_mse(lcx, lcy, rcx, rcy, co):
    w, h, c = sample2.shape
    nr,  = lcx.shape
    
    test = np.zeros((w, h, c), np.uint8)
    test[:, :] = (255, 255, 255)
    
    for i in range(nr):
        test[lcy[i]:rcy[i], lcx[i]:rcx[i]] = 0.5 * test[lcy[i]:rcy[i], lcx[i]:rcx[i]] + 0.5* co[i]
    
    err = 100*np.sum((test.astype("float") - sample2.astype("float"))**2)
    mse_list.append(err)
    return np.array(err, dtype=np.int64)
```

```python
@as_op(itypes=[tt.lvector, tt.lvector, tt.lvector, tt.lvector, tt.lmatrix], otypes=[tt.lscalar])
def compute_ssim(lcx, lcy, rcx, rcy, co):
    w, h, c = sample2.shape
    nr,  = lcx.shape
    
    test = np.zeros((w, h, c), np.uint8)
    test[:, :] = (255, 255, 255)
    
    for i in range(nr):
        test[lcy[i]:rcy[i], lcx[i]:rcx[i]] = 0.5 * test[lcy[i]:rcy[i], lcx[i]:rcx[i]] + 0.5* co[i]

    err = 100 * ssim(test, sample2, data_range=test.max() - test.min(), multichannel = True)
    ssim_list.append(err)
    return np.array(err, dtype=np.int64)
```

```python
# compute_mse 함수를 조금 수정하여 적용해보았습니다. 결과상 차이는 없는것으로 보이고, computation time이 조금 줄어드는 것 같습니다.

num_rect = 500
h, w, c = sample2.shape

with pm.Model() as model:
    # Priors
    left_corner_x = pm.DiscreteUniform('left_corner_x', lower=0, upper=w, shape=num_rect)
    left_corner_y = pm.DiscreteUniform('left_corner_y', lower=0, upper=h, shape=num_rect)
    right_corner_x = pm.DiscreteUniform('right_corner_x', lower=0, upper=w, shape=num_rect)
    right_corner_y = pm.DiscreteUniform('right_corner_y', lower=0, upper=h, shape=num_rect)

    color = pm.DiscreteUniform('color', lower=0, upper=255, shape=(num_rect, 3))

    #obj = rec_KL(left_corner_x, left_corner_y, right_corner_x, right_corner_y, color)
    #obj = compute_ssim(left_corner_x, left_corner_y, right_corner_x, right_corner_y, color)
    obj = compute_mse(left_corner_x, left_corner_y, right_corner_x, right_corner_y, color)
    
    # Data likelihood
    objective_obs = pm.Normal('objective_obs', mu=obj, sd=1, observed=0)
    #objective_obs = pm.Poisson('objective', obj, observed=0)

    step = pm.Metropolis()

    # Initial values for stochastic nodes
    start = {'left_corner_x': np.random.randint(w, size=num_rect), 'left_corner_y': np.random.randint(h, size=num_rect),
             'right_corner_x': np.random.randint(w, size=num_rect), 'right_corner_y': np.random.randint(h, size=num_rect),
             'color': np.random.randint(255, size=(num_rect, 3))}

    tr = pm.sample(10000, start=start, step=[step], cores=1)
    #pm.traceplot(tr)

#print(pm.summary(tr))
plt.show()
```
